```python
import numpy as np
import matplotlib.pyplot as plt
x=np.linspace(-5,5)
y_1=2*x
y_2=3*x

#축의 라벨
plt.xlabel("x value")
plt.ylabel("y value")

#그래프의 타이틀
plt.title("My graph")

#플롯 범례와 선 스타일을 지정
plt.plot(x,y_1,label="y1")
plt.plot(x,y_2,label="y2",linestyle="dashed")
plt.legend()

plt.show()
```

```python
import pandas as pd
import numpy as np
```

```python
a=pd.Series([60,80,70,50,30],index=["Korean","English","Math","Science","History"])
print(type(a))
print(a)
```

numpy를 이용해 데이터 지정

```python
a=pd.Series(np.array([60,80,70,50,30]),index=np.array(["Korean","English","Math","Science","History"]))
print(type(a))
print(a)
```

사전을 사용해 데이터 삽입


```python
a=pd.Series({"Korean" : 60, "English" : 80,"Math" : 70,"Science": 50,"History": 30})
print(type(a))
print(a)
```

데이터 접근 방법

```python
a=pd.Series([60,80,70,50,30],index=["Korean","English","Math","Science","History"])
print(type(a))
print(a[2])
print(a["Math"])
```

append메소드를 사용해서 추가 가능 (데이터 변경 및 삭제 Series끼리 결합도 가능)

```python
a=pd.Series([60,80,70,50,30],index=["Korean","English","Math","Science","History"])
b=pd.Series([20],index=["Arts"])
a=a.append(b)
print(a)
```
