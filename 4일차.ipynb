인공뉴런 : 뉴런에는 여러개의 입력이 있지만 출력은 1개다.
 각 입력에는 가중치(중요도)를 곱한다. 각 입력마다 가중치가 다르므로 가중치의 값이 뇌의 시냅스에서의 전달 효율에 상응하며, 값이 크면 그만큼 많은 정보가 흘러 나온다. 그리고나서 가중치와 입력을 곱한 값의 총합에 바이어스라고 하는 상수를 더한다.
*바이어스 : 뉴런의 감도
입력과 가중치의 곱의 총합에 바이어스를 더한 값은 활성화 함수로 처리된다. 이 함수는 입력을 뉴런의 비선형 상태 신호로 변환한다. 이 뉴런을 연결해서 구축한 망이 신경망이다.

* (입력*가중치)++입력들 +바이어스(상수) -> 활성화 함수 -> 신경망 -> 출력

신경망
-여러개의 입력과 여러개의 출력이 있음
-수치를 입력하고, 정보를 전파시켜 결과를 출력
-출력은 확률 등의 예측치로 해석할 수 있고, 망에 의해 예측할수 있음.
-간단한 기능만 가진 뉴런이 층을 형성하고, 층 사이에서 접속이 이뤄지면서 형성

역전파
-신경망은 출력과 정답의 오차가 작아지도록 가중치와 바이어스를 조정해서 학습 가능
-1층씩 거슬러 올라가듯이 오차를 전파시켜 가중치와 바이어스를 갱신하는데 이 알고리즘을 사용
-역전파에서 신경망에 데이터가 거슬러 올라가도록해 망의 각 층의 파라미터가 조정
-각 파라미터가 반복해 조정됨으로써 망은 점차 학습하고, 적절한 예측이 이루어짐

딥러닝
-다수의 층으로 이루어진 신경망학습
-기본적으로 층의 수가 많아질수록 망의 표현력은 향상되지만, 그에 따라 학습은 어려워짐

```python
import numpy as np
from sklearn import datasets    #scikit-learn이라고 하는 라이브러리에서 Iris dataset을 읽어들임

iris=datasets.load_iris()
print(iris.data[:10]) #4개의 측정값을 10세트 표시
print(iris.data.shape)  #측정값 데이터의 형태를 표시
print(iris.target)    #라벨을 전부 표시
```

위의 출력한 마지막데이터에서 0,1,2의 3종류가 있는것을 알수 있고, 이것은 신경망을 훈련하기 위해서 사용하는 정답데이터가 됨.

```python
import numpy as np
import np_utils
from sklearn import preprocessing
from keras.utils import np_utils

scaler=preprocessing.StandardScalers()  #표준화하기 위한 스케일러
scaler.fit(iris.data)   #변환하기 위한 파라미터를 계산
x=scaler.transform(iris.data) #데이터 변환
print(x[:10])

#정답 데이터
t=np_utils.to_categorical(iris.target)  #라벨을 one-hot 표현으로 변환
print(t[:10])   #정답라벨을 10건 표시
```

```python
from sklearn import preprocessing
from keras.utils import np_utils

# ---- 입력 데이터 ----
scaler = preprocessing.StandardScaler()  # 표준화하기 위한 스케일러
scaler.fit(iris.data)  # 변환하기 위한 파라미터를 계산
x = scaler.transform(iris.data)  # 데이터 변환
print(x[:10])  # 입력을 10건 표시

# ---- 정답 데이터  ----
t = np_utils.to_categorical(iris.target)  # 라벨을 one-hot 표현으로 변환
print(t[:10])  # 정답 라벨을 10건 표시
```
